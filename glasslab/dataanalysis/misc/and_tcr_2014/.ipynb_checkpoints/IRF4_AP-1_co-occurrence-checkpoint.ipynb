```python
from __future__ import division
import pandas
from glasslab.dataanalysis.motifs.motif_analyzer import MotifAnalyzer
import subprocess
import numpy as np
from scipy.stats import chi2
```

```python
def find_occurrence_of(motifs, peak_file, output_filename, size=False, histogram=False):
    yzer = MotifAnalyzer()
    
    dirpath = 'karmel/Desktop/Projects/GlassLab/Notes_and_Reports/AND_TCR/Analysis/Chips1_2/Motifs'
    dirpath = yzer.get_path(dirpath)
    for ab in ('me2', 'ac'):
        for peptide in ('K99A','NoPep','PCC'):
            pep_dirpath = yzer.get_filename(dirpath, 
                            '{}_{}'.format(peptide, ab))
            
            filename = yzer.get_filename(pep_dirpath, peak_file)
        
            output_path = yzer.get_and_create_path(pep_dirpath, 'motifs_of_interest')
            output_path = yzer.get_and_create_path(output_path, '_'.join(motifs))
            output_file = yzer.get_filename(output_path, output_filename)
            print output_file
            motif_files = []
            for motif in motifs:
                motif_files.append(yzer.get_filename(dirpath, 'motifs_of_interest', (motif + '.motif')))
            motif_string = '" "'.join(motif_files)
            
            hist_string = ''
            if histogram:
                hist_string = '-hist 10'
                if not size: size = 1200
                    
            size_string = size and '-size {}'.format(size) or ''
            
            command = '{}annotatePeaks.pl "{}" mm9 -m "{}" {} {} > "{}"'.format(
                                            yzer.homer_bin, filename,
                                            motif_string, size_string,
                                            hist_string, output_file)

            subprocess.check_call(command, shell=True)
```

```python
def chi_squared(motifs, filename, columns):
    yzer = MotifAnalyzer()
    
    dirpath = 'karmel/Desktop/Projects/GlassLab/Notes_and_Reports/AND_TCR/Analysis/Chips1_2/Motifs'
    dirpath = yzer.get_path(dirpath)
    for ab in ('me2', 'ac'):
        for peptide in ('K99A','NoPep','PCC'):
            pep_dirpath = yzer.get_filename(dirpath, 
                            '{}_{}'.format(peptide, ab))
            
            annotated_peaks = yzer.get_and_create_path(pep_dirpath, 'motifs_of_interest')
            annotated_peaks = yzer.get_and_create_path(annotated_peaks, '_'.join(motifs))
            annotated_peaks = yzer.get_filename(annotated_peaks, filename)
            
            data = pandas.read_csv(annotated_peaks, header=0, sep='\t')
            
            subset = data[columns]
            subset = subset.fillna(0)
            
            # Convert to ints
            for col in subset.columns:
                subset[col] = subset[col].map(lambda x: int(bool(x)))
                
            # Get counts of intersections
            counts = {}
            counts['None'] = len(subset[subset.sum(axis=1) <= 0])
            for col in subset.columns:
                counts[col] = len(subset[subset[col] == 1])
            counts['All'] = len(subset[subset[columns].sum(axis=1) >= len(subset.columns)])

            print peptide, ab
            print counts
            
            # Chi-squared stat
            # @todo: this shouldn't be hard-coded...
            total = len(subset)
            first_only = counts[columns[0]] - counts['All']
            first_no = total - counts[columns[0]]
            second_only = counts[columns[1]] - counts['All']
            second_no = total - counts[columns[1]]
            
            observed = np.array([[counts['None'], first_only],
                                 [second_only, counts['All']]])
            expected = np.array([[(first_no*second_no)/total, 
                                  (counts[columns[0]]*second_no)/total],
                                 [(first_no*counts[columns[1]])/total, 
                                  (counts[columns[0]]*counts[columns[1]])/total]])
            
            print 'Observed:', observed
            print 'Expected:', expected
            
            obs_minus_exp = observed - expected
            chi_sq = sum(obs_minus_exp**2/expected)
            print 'Chi-squared:', chi_sq
            
            print 'p-value:', chi2.pdf(chi_sq, df=1)*2
```

```python
cols = ['IRF4(IRF)/GM12878-IRF4-ChIP-Seq(GSE32465)/Homer Distance From Peak(sequence,strand,conservation)',
        'AP-1(bZIP)/ThioMac-PU.1-ChIP-Seq(GSE21512)/Homer Distance From Peak(sequence,strand,conservation)']
chi_squared(['irf4','ap1'], 'irf4_ap1_occurrence_in_all_peaks.txt',
               cols)


```
