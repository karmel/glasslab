```python
from __future__ import division
from glasslab.dataanalysis.graphing.seq_grapher import SeqGrapher
import pandas
```

```python
peptide = 'NoPep'
ab = 'ac'
```

```python
yzer = SeqGrapher()
dirpath = 'karmel/Desktop/Projects/GlassLab/Notes_and_Reports/AND_TCR/Analysis/Chips1_2/Motifs'
dirpath = yzer.get_path(dirpath)
```

```python
pep_dirpath = yzer.get_filename(dirpath, '{}_{}'.format(peptide, ab))
filename = yzer.get_filename(pep_dirpath, '{}_{}_enhancers_batf.txt'.format(peptide, ab))
```

```python
data = yzer.import_file(filename)
data = data.fillna(0)
```

```python
length = 2000
data = data[(data['length'] < length)]
```

```python
with_batf = data[data['batf_tag_count'] > 0]
with_ets1 = data[data['ets1_tag_count'] > 0]
with_batf_ets1 = with_batf[with_batf['ets1_tag_count'] > 0]

print('Total {} {} peaks'.format(peptide, ab), len(data))
print('With Batf', len(with_batf))
print('With Ets1', len(with_ets1))
print('With Batf+Ets1', len(with_batf_ets1))
```

```python
print('Total {} {} peaks'.format(peptide, ab), len(data)/len(data))
print('With Batf', len(with_batf)/len(data))
print('With Ets1', len(with_ets1)/len(data))
print('With Batf+Ets1', len(with_batf_ets1)/len(data))
```

```python
with_batf_only = len(with_batf) - len(with_batf_ets1)
with_ets1_only = len(with_ets1) - len(with_batf_ets1)
with_both = len(with_batf_ets1)
leftover = len(data) - with_batf_only - with_ets1_only - with_both
yzer.piechart([with_batf_only, with_both, with_ets1_only, leftover],
              ['Has Batf', 'Has Batf + Ets1', 'Has Ets1', 'Has neither'],
              title = '{} {} Peaks < {} bp overlap\nwith Batf in Th17 Cells\nand Ets1 in naive CD4'.format(peptide, ab, length),
              colors = ['#FFABAB','#D9ABFF','#ABE4FF','#CCCCCC'],
              save_dir=yzer.get_filename(dirpath,'../Figures/CD4_Ets1_pies'))
```

```python
pep_only = data[(data['k99a_tag_count'] == 0) & (data['pcc_tag_count'] == 0)]
pep_only_with_batf = pep_only[pep_only['batf_tag_count'] > 0]
pep_only_with_ets1 = pep_only[pep_only['ets1_tag_count'] > 0]
pep_only_with_batf_ets1 = pep_only_with_batf[pep_only_with_batf['ets1_tag_count'] > 0]

print('{} {} only peaks'.format(peptide, ab), len(pep_only))
print('With Batf', len(pep_only_with_batf))
print('With Ets1', len(pep_only_with_ets1))
print('With Batf+Ets1', len(pep_only_with_batf_ets1))
```

```python
print('{} {} only peaks'.format(peptide, ab), len(pep_only)/len(data))
print('With Batf', len(pep_only_with_batf)/len(pep_only))
print('With Ets1', len(pep_only_with_ets1)/len(pep_only))
print('With Batf+Ets1', len(pep_only_with_batf_ets1)/len(pep_only))
```

```python
pep_only_with_batf_only = len(pep_only_with_batf) - len(pep_only_with_batf_ets1)
pep_only_with_ets1_only = len(pep_only_with_ets1) - len(pep_only_with_batf_ets1)
pep_only_with_both = len(pep_only_with_batf_ets1)
pep_only_leftover = len(pep_only) - pep_only_with_batf_only \
                    - pep_only_with_ets1_only - pep_only_with_both
yzer.piechart([pep_only_with_batf_only, pep_only_with_both, 
               pep_only_with_ets1_only, pep_only_leftover],
              ['Has Batf', 'Has Batf + Ets1', 'Has Ets1', 'Has neither'],
              title = '{} {} Only Peaks < {} bp overlap\nwith Batf in Th17 Cells\nand Ets1 in naive CD4'.format(peptide, ab, length),
              colors = ['#FFABAB','#D9ABFF','#ABE4FF','#CCCCCC'],
              save_dir=yzer.get_filename(dirpath,'../Figures/CD4_Ets1_pies'))
```

```python
leftover = len(data) - len(with_ets1)
yzer.piechart([len(with_ets1), leftover],
              ['Has Ets1', 'Has neither'],
              title = '{} {} Peaks < {} bp overlap\nwith Ets1 in naive CD4'.format(peptide, ab, length),
              colors = ['#ABE4FF','#CCCCCC'],
              save_dir=yzer.get_filename(dirpath,'../Figures/CD4_Ets1_pies'))
```
