```python
schema = 'motifs'
motif = 'ets1'
def print_sql(s):
    for ab in ('me2', 'ac'):
        for peptide in ('no_pep', 'k99a', 'pcc'):
            print(s.format(schema, motif, peptide, ab))
```

```python

s = '''CREATE TABLE "{0}"."{1}_{2}_{3}" (
	"id" int4 NOT NULL DEFAULT NULL,
	"chromosome_id" int4 DEFAULT NULL,
	"chrom" varchar(20) NOT NULL DEFAULT NULL,
	"start" int8 NOT NULL DEFAULT NULL,
	"end" int8 NOT NULL DEFAULT NULL,
	"strand" int2 DEFAULT NULL,
	"start_end" int8range DEFAULT NULL,
	"score" float4 NOT NULL DEFAULT NULL,
	"strand_char" varchar(1) NOT NULL DEFAULT NULL,
	CONSTRAINT "{1}_{2}_{3}_pkey" PRIMARY KEY ("id")
)
WITH (OIDS=FALSE);
ALTER TABLE "{0}"."{1}_{2}_{3}" OWNER TO "vespucci_user";
CREATE INDEX "{1}_{2}_{3}_chr_id_idx" ON "{0}"."{1}_{2}_{3}" USING btree(chromosome_id ASC NULLS LAST);
CREATE INDEX "{1}_{2}_{3}_start_end_idx" ON "{0}"."{1}_{2}_{3}" USING btree(start_end ASC NULLS LAST);


CREATE SEQUENCE "{0}"."{1}_{2}_{3}_id_seq"
	START WITH 1
	INCREMENT BY 1
	NO MINVALUE
	NO MAXVALUE
	CACHE 1;
ALTER SEQUENCE "{0}"."{1}_{2}_{3}_id_seq" OWNED BY "{0}"."{1}_{2}_{3}".id;
ALTER TABLE "{0}"."{1}_{2}_{3}" ALTER COLUMN id 
	SET DEFAULT nextval('"{0}"."{1}_{2}_{3}_id_seq"'::regclass);
ALTER TABLE ONLY "{0}"."{1}_{2}_{3}" ADD CONSTRAINT {1}_{2}_{3}_pkey PRIMARY KEY (id);'''

print_sql(s)
```

```python
s = '''
        Update "{0}"."{1}_{2}_{3}" t
            SET chromosome_id = chr.id,
            strand = (CASE WHEN t.strand_char = '-' THEN 1 ELSE 0 END),
            start_end = int8range(t."start", t."end", '[]')
        FROM "genome_reference_mm9"."chromosome" chr
        WHERE t."chrom" = chr."name";'''
print_sql(s)
```

```python
s = '''
ALTER TABLE "{0}"."{1}_{2}_{3}" DROP COLUMN chrom;
ALTER TABLE "{0}"."{1}_{2}_{3}" DROP COLUMN strand_char;'''

print_sql(s)
```
